```python
def payment_generator(payment, n=48):
    
    
    if payment['freq']=='M':
        df = pd.DataFrame(data={'dates': month_range(payment['start_date'], n)})
    else:
        df = pd.DataFrame(data={'dates': pd.date_range(payment['start_date'], periods=n, freq=payment['freq'])})
    
    df['account_name'] = payment['account_name'] 
    df['account_IBAN'] = payment['account_IBAN']
    df['other_account_name'] = payment['other_account_name']
    df['Type'] = payment['Type']
    
    if payment['amount']['variable'] == 'False':
        df['amount'] = payment['amount']['base']
    
    if payment['amount']['variable'] == 'True':
        df['amount'] = payment['amount']['base']
        df['amount'] = df['amount'].apply(lambda x: x + random.uniform(payment['amount']['var'], 0))
    
    return df
```

```python
def month_range(start, periods=12):
    rng = pd.date_range(pd.Timestamp(start)-pd.offsets.MonthBegin(),
                        periods=periods,
                        freq='MS')
    ret = (rng + pd.offsets.Day(pd.Timestamp(start).day-1)).to_series()
    ret.loc[ret.dt.month > rng.month] -= pd.offsets.MonthEnd(1)
    return pd.DatetimeIndex(ret)
```

```python
def payment_episode(payment):
    df = pd.DataFrame(data={'dates': pd.date_range(payment['start_date'], periods=48, freq="7D")})
    date = df.loc[random.randint(0,df.shape[0]-1)]
    num_payments = random.randint(5,12)
    df = df[df['dates'] == date['dates']]
    df = df.loc[df.index.repeat(num_payments)]
    df['account_name'] = payment['account_name'] 
    df['account_IBAN'] = payment['account_IBAN']
    df['other_account_name'] = payment['other_account_name']
    df['amount'] = payment['amount']['base']
    df['amount'] = df['amount'].apply(lambda x: x + random.uniform(payment['amount']['var'], 0))
    return df
```

```python
list_impluse_episode = []
for x in payments_impluse:
    epi = payment_episode(x)
    list_impluse_episode.append(epi)
df_impluse_episode = pd.concat(list_impluse_episode)
df_impluse_episode['dates'] = df_impluse_episode['dates'].dt.strftime('%Y-%m-%d %H:%M:%S')
df_impluse_episode['time_offset'] =  random.randint(0,1220)
df_impluse_episode['time_offset'] = df_impluse_episode['time_offset'].apply(lambda x: x + random.uniform(0, 11))
df_impluse_episode['dates'] = pd.to_datetime(df_impluse_episode['dates']) + pd.to_timedelta(df_impluse_episode['time_offset'], unit='m')
```
